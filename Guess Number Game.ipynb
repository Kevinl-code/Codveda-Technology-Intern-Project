```python
import random

def game():
    # Generates a random number between 1 and 100
    num = random.randint(1, 100)
    
    # Set maximum number of attempts
    attempts = 3
    attempt = 0
    
    print("Number Guessing Game!")
    print(f"You have {attempts} attempts to guess it.\n")
    
    while attempt < attempts:
        try:
            # Get user input
            guess = int(input("Enter your guess: "))
            attempt += 1
            
            # Check if the guess is correct
            if guess == num:
                print(f"\nCongratulations! You guessed the number {num} correctly!")
                print(f"It took you {attempt} attempt(s).")
                return
            
            # Provide feedback
            elif guess < num:
                print("Too low!")
            else:
                print("Too high!")
            
            # Show remaining attempts
            remains = attempts - attempt
            if remains > 0:
                print(f"You have {remains} attempt(s) left.\n")
            else:
                print()  
                
        except ValueError:
            print("Please enter a valid number!")
            attempt -= 1 
    
    print(f"Game Over! You've used all {attempts} attempts.")
    print(f"The number I was thinking of was: {num}")

# Run the game
if __name__ == "__main__":
    game()
```
