```python
# Original-style function with single-line step-by-step comments
def wc():
    # ask the user to type the filename (e.g., "sample.txt")
    filename = input("Enter filename: ")
    try:
        # open the file for reading; this ensures the file is closed automatically
        with open(filename, 'r', encoding='utf-8') as file:
            # read the entire file content into a string variable
            text = file.read()
            # print the full text to confirm the file was read successfully
            print(f"Text Contains:\n{text}")

        # open the same file again to count words (keeps your original structure)
        with open(filename, 'r', encoding='utf-8') as file:
            # read the file and split on whitespace to produce a list of words
            words = file.read().split()
            # print the length of the words list which is the word count
            print(f"Word count: {len(words)}")

    # handle the case where the filename given by the user does not exist
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    # catch-all for any other unexpected errors and show a simple message
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
wc()
```
