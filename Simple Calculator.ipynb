```python
# Read the first number from the user, convert to an integer
a = int(input("Enter the First Number: "))

# Read the second number from the user, convert to an integer
b = int(input("Enter the Second Number: "))

# Read the operator choice from the user as a string
c = input("Choose your choice (+, -, *, /): ")

# Define add function: returns the sum of two numbers
def add(a, b):
    return a + b

# Define subtract function: returns the difference a - b
def sub(a, b):
    return a - b

# Define multiply function: returns the product of two numbers
def mul(a, b):
    return a * b

# Define divide function:
# - If b is zero, return an error string to avoid ZeroDivisionError
# - Otherwise return a / b (floating-point division)
def div(a, b):
    if b == 0:
        return "Cannot divide by zero"
    else:
        return a / b

# Decide which operation to run based on user input 'c'
if c == "+":
    result = add(a, b)
elif c == "-":
    result = sub(a, b)
elif c == "*":
    result = mul(a, b)
elif c == "/":
    result = div(a, b)
else:
    # If operator is none of the allowed ones, mark result as invalid operator
    result = "Invalid operator"

# Print the final result to the console
print("Result:", result)
```
