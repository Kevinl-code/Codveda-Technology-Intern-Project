```python
# Import requests beautifulsoup4 csv

import requests  
from bs4 import BeautifulSoup  
import csv  

url = "https://www.bbc.com/news"  # target web page URL to scrape headlines
filename = "headlines.csv"  # name of the CSV file where headlines will be saved

try:
    response = requests.get(url)  # send an HTTP GET request to the URL and store the response
    response.raise_for_status()  # raise an exception if the response status indicates an error
    soup = BeautifulSoup(response.text, "html.parser")  # parse the HTML content using BeautifulSoup

    # find all <h3> tags, extract their text, strip whitespace, and keep only non-empty strings
    headlines = [h3.get_text().strip() for h3 in soup.find_all("h3") if h3.get_text().strip()]

    with open(filename, "w", newline="", encoding="utf-8") as f:  # open the CSV file for writing with UTF-8 encoding
        writer = csv.writer(f)  # create a CSV writer object to write rows into the file
        writer.writerow(["Headline"])  # write the CSV header row with a single column "Headline"
        for h in headlines:  # iterate over the collected headlines
            writer.writerow([h])  # write each headline as a new row in the CSV file

    print(f"Scraped {len(headlines)} headlines and saved to {filename}")  # print a summary of the result

except requests.exceptions.RequestException as e:  # catch network-related errors from requests
    print("Network error:", e)  # print the error message for debugging
```
